## Time since last project in the same category
code from:
https://www.kaggle.com/matleonard/feature-generation

```python
def time_since_last_project(series):
    # Return the time in hours
    return series.diff().dt.total_seconds() / 3600.

data = rdata[['category', 'launched']].sort_values('launched')
timedeltas = data.groupby('category').transform(time_since_last_project)
timedeltas.head(20)
```

```python
# Final time since last project
timedeltas = timedeltas.fillna(timedeltas.median()).reindex(rdata.index)
timedeltas.head(20)
```

# Exploration

```python
rdata["launched"] = timedeltas
rdata.dropna(inplace = True)
rdata.shape
```

Number of projects in last week

code from: https://www.kaggle.com/matleonard/feature-generation

```python
rdata = pd.read_csv("../Data/Cleaned/2018.csv", parse_dates=['launched'])
```

```python
launched = pd.Series(rdata.index, index=rdata.launched, name="count_7_days").sort_index()
launched.head(20)
```

```python
count_7_days = launched.rolling('7d').count() - 1
print(count_7_days.head(20))

# Ignore records with broken launch dates
plt.plot(count_7_days[7:]);
plt.title("Number of projects launched over periods of 7 days");
```

```python
count_7_days.index = launched.values
count_7_days = count_7_days.reindex(rdata.index)
```

```python
rdata = rdata.join(count_7_days).head(10)
```

```python
rdata["count_7_days"].where(rdata["state"] == 1).mean()
```

```python
rdata["count_7_days"].where(rdata["state"] == 0).mean()
```
