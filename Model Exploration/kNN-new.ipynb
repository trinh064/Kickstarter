```python
# Importing libraries
# we don't like warnings; you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
from matplotlib import pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

```python
# Importing data
df_ks = pd.read_csv("../Data/Cleaned/joined_data.csv")
```

```python
# Trying model off the shelf
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
scaler = StandardScaler()

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
```

## Encoding Categorical Data

```python
# Encoding Category
df_ks = pd.get_dummies(df_ks, columns=['category'], prefix=['category'])
```

```python
# Encoding Main_Category -- This is redundant, so remove if slow
df_ks = pd.get_dummies(df_ks, columns=['main_category'], prefix=['main_category'])
```

```python
# Encoding location_state
df_ks = pd.get_dummies(df_ks, columns=['location_state'], prefix=['loc_state'])
```

```python
# Encoding country
df_ks = pd.get_dummies(df_ks, columns=['country'], prefix=['country'])
```

```python
# Remove Text-based Description
df_ks = df_ks.drop(columns=['blurb','name'])
```

```python
knn = KNeighborsClassifier()
#knn = KNeighborsClassifier(n_neighbors=5)
```

```python
import timeit
import time
import datetime
```

```python
num_to_test = 40000
max_train_time = 600

start = timeit.default_timer()

stop = timeit.default_timer()

df_ks_full = df_ks.copy()

df_ks = df_ks.sample(num_to_test)
feature_names = df_ks.drop(columns=['state','duration_days']).columns
X_test = df_ks[feature_names].copy()
X_test_scaled = scaler.fit_transform(X_test)
y_test = df_ks.state

num_samples = 8

num_samples_list = []
accuracy_list = []
time_list = []
```

## Baseline Model -- Hand Picked
Can't do it on all features in the interest of time

```python
print(f"Testing on {num_to_test} samples")
print(f"Max train-test time: {max_train_time/60} minutes\n")

while stop - start < max_train_time and num_samples < df_ks_full.shape[0]:
    print(f"{num_samples} training samples")
    
    df_ks = df_ks_full.sample(num_samples, replace = True)
    
    start = timeit.default_timer()

    # Looking at all features (except for text, and generated)
    feature_names = df_ks.drop(columns=['state','duration_days']).columns
    X_train = df_ks[feature_names]
    y_train = df_ks.state
    
    # for kNN, we need to scale features
    X_train_scaled = scaler.fit_transform(X_train)

    knn.fit(X_train_scaled, y_train)

    knn_pred = knn.predict(X_test_scaled)
    stop = timeit.default_timer()
    
    acc = accuracy_score(y_test, knn_pred)
    
    print (f" Accuracy: {acc}")
    print(f" Time: {str(datetime.timedelta(seconds=stop-start))}\n")
    
    num_samples_list.append(num_samples)
    accuracy_list.append(acc)
    time_list.append(round(stop-start, 3))
    
    num_samples *= 2
    time.sleep(1)
```

```python
plt.plot(num_samples_list, accuracy_list)
plt.ylabel('Accuracy')
plt.xlabel('Number of Training Samples')
plt.show()
```

```python
plt.plot(num_samples_list, time_list)
plt.ylabel('Time (sec)')
plt.xlabel('Number of Training Samples')
plt.show()
```
